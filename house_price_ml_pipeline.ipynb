```python
nominal_cols = [col for col in cat_cols if col not in ordinal_cols and col not in binary_cols]
nominal_cols
```

```python
for col in binary_cols:
    X_train[col] = X_train[col].map({'Y':1, 'N':0})
    X_test[col] = X_test[col].map({'Y':1, 'N':0})
```

```python
X_train[binary_cols].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Initialize encoder (UPDATED)
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# Fit only on training data
ohe.fit(X_train[nominal_cols])

# Transform both train and test
X_train_ohe = ohe.transform(X_train[nominal_cols])
X_test_ohe  = ohe.transform(X_test[nominal_cols])

# Get feature names
ohe_cols = ohe.get_feature_names_out(nominal_cols)

# Convert to DataFrame
X_train_ohe = pd.DataFrame(X_train_ohe, columns=ohe_cols, index=X_train.index)
X_test_ohe  = pd.DataFrame(X_test_ohe, columns=ohe_cols, index=X_test.index)

# Drop original columns
X_train = X_train.drop(columns=nominal_cols)
X_test  = X_test.drop(columns=nominal_cols)

# Concatenate encoded features
X_train = pd.concat([X_train, X_train_ohe], axis=1)
X_test  = pd.concat([X_test, X_test_ohe], axis=1)
```

```python
qual_map = {
    'None': 0,
    'Po': 1,
    'Fa': 2,
    'TA': 3,
    'Gd': 4,
    'Ex': 5
}

for col in ordinal_cols:
    if ('Qual' in col) or ('Cond' in col) or ('QC' in col):
        X_train[col] = X_train[col].map(qual_map)
        X_test[col]  = X_test[col].map(qual_map)
```

```python
# Garage Finish
X_train['Garage Finish'] = X_train['Garage Finish'].map({'None':0, 'Unf':1, 'RFn':2, 'Fin':3})
X_test['Garage Finish']  = X_test['Garage Finish'].map({'None':0, 'Unf':1, 'RFn':2, 'Fin':3})

# Land Slope
X_train['Land Slope'] = X_train['Land Slope'].map({'Gtl':1, 'Mod':2, 'Sev':3})
X_test['Land Slope']  = X_test['Land Slope'].map({'Gtl':1, 'Mod':2, 'Sev':3})

# Lot Shape
X_train['Lot Shape'] = X_train['Lot Shape'].map({'IR3':1, 'IR2':2, 'IR1':3, 'Reg':4})
X_test['Lot Shape']  = X_test['Lot Shape'].map({'IR3':1, 'IR2':2, 'IR1':3, 'Reg':4})

# Functional
X_train['Functional'] = X_train['Functional'].map({
    'Sal':1, 'Sev':2, 'Maj2':3, 'Maj1':4,
    'Mod':5, 'Min2':6, 'Min1':7, 'Typ':8
})
X_test['Functional']  = X_test['Functional'].map({
    'Sal':1, 'Sev':2, 'Maj2':3, 'Maj1':4,
    'Mod':5, 'Min2':6, 'Min1':7, 'Typ':8
})
```

```python
X_train[ordinal_cols].isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=0)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_test, y_pred)
rmse = rmse ** 0.5
rmse
```
